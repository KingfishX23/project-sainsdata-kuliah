```python
import pandas as pd
import numpy as np

#* open file csv
dataFrameTwitter = pd.read_csv('TRANSLATED-covid-sentiment.csv')

#* ilangin kolom2 berikut
columns = ['mentions', 'conversation_id', 'user_id', 'hashtags']
dataFrameTwitter.drop(columns, inplace=True, axis=1)


"""
#* Convert date and time column to unix timestamp
# This Process converts, and then delete the old date and time columns
# in exchange for a new column (timestamps)
"""
# Declare variable timeStamps untuk ngegabungin dua kolom tersebut
# "2020-05-07 23:58:29"
timeStamps = dataFrameTwitter['date'] + " " + dataFrameTwitter['time']

# Convert String menjadi bentuknya integer!

dataFrameTwitter['timeStamps']= pd.to_datetime(timeStamps)
```

```python
dataFrameTwitter.sort_values(by="retweets_count", ascending=False)
```

```python
dataFrameTwitter.sort_values(by="likes_count", ascending=False)
```

```python
dataFrameTwitter.sort_values(by="replies_count", ascending=False)
```

```python
dataFrameTwitter['dateofweek'] = dataFrameTwitter['timeStamps'].dt.day_name()
dataFrameTwitter['timeofday'] = dataFrameTwitter['timeStamps'].dt.hour
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

"""
#* Mencoba COUNTIF dan COUNTIFS di kolom likes_count yang nilainya diantara 0 - 5
# Setara dengan di excel: COUNTIFS(XX:YY, kondisi1, kondisi2, ..., kondisi_n) !!
"""

#Tarik data apabila kondisinya data sama dengan 0
#Jumlahin semua yang memenuhi kondisi tersebut

daySunday = (dataFrameTwitter["dateofweek"] == "Sunday") 
totalSunday = np.sum(daySunday) 

dayMonday = (dataFrameTwitter["dateofweek"] == "Monday") 
totalMonday = np.sum(dayMonday)

dayTuesday = (dataFrameTwitter["dateofweek"] == "Tuesday") 
totalTuesday = np.sum(dayTuesday)

dayWednesday = (dataFrameTwitter["dateofweek"] == "Wednesday") 
totalWednesday = np.sum(dayWednesday)

dayThursday = (dataFrameTwitter["dateofweek"] == "Thursday") 
totalThursday = np.sum(dayThursday)

dayFriday = (dataFrameTwitter["dateofweek"] == "Friday") 
totalFriday = np.sum(dayFriday)

daySaturday = (dataFrameTwitter["dateofweek"] == "Saturday") 
totalSaturday = np.sum(daySaturday)


plotDf = pd.DataFrame(
    [["Sun",totalSunday],["Mon",totalMonday],["Tue",totalTuesday],["Wed",totalWednesday],["Thu",totalThursday],["Fri",totalFriday],["Sat",totalSaturday]], #kolom
    columns=["Day","Tweet Number"]
)

sns.barplot(x="Day", y="Tweet Number", data=plotDf)
```
